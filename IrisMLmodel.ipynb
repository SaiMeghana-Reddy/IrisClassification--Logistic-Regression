```python
#Import necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#Load the dataset
iris_data=pd.read_csv("Iris.csv")
iris_data.head()
```

```python
#featues
X=iris_data.drop(["Id","Species"],axis=1)
#Label
y=iris_data["Species"]
```

```python
#Create and train the model
model=LogisticRegression()
model.fit(X.values,y)
```

```python
#make predictions
new_data=[[5.1,3.5,1.4,0.2]]
Prediction=model.predict(new_data)
print(Prediction[0])
```

```python
import matplotlib.pyplot as plt
# Scatter plot
plt.figure(figsize=(8,6))

for species in iris_data['Species'].unique():
    subset = iris_data[iris_data['Species'] == species]
    plt.scatter(subset['SepalLengthCm'], subset['SepalWidthCm'], label=species)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Width (Iris Dataset)")
plt.legend()
plt.grid()

plt.show()
```
